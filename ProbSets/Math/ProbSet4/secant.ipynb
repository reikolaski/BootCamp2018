```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Use the secant method to find the minimizer of a function
def secant_minimizer(x0, x1, ϵ):
    fig, ax = plt.subplots(figsize=(15,8))
    
    max_iter = 1000
    x = np.zeros(max_iter, dtype=float)
    
    # initial guesses
    x[0] = x0
    ax.plot(x[0], f(x[0]), "o", markersize=10)
    x[1] = x1
    ax.plot(x[1], f(x[1]), "o", markersize=10)
    
    k = 2
    dist = 1e3
    while dist > ϵ and k < max_iter:
        x[k+1] = x[k] - fprime(x[k]) * ((x[k] - x[k-1]) / (fprime(x[k]) - fprime(x[k-1])))  # next approximation
        dist = np.abs((x[k+1] - x[k]) / x[k])  # check distance from previous iteration
        
        ax.plot(x[k+1], f(x[k+1]), "o", markersize=10)
        k +=1
    
    x_interval = np.linspace(x[0], x[k] + 1, 100)
    ax.plot(x_interval, f(x_interval), 'k')
    
    plt.show()
    
    return x[k]
```

```python
f = lambda x: x ** 4 - 14 * x ** 3 + 60 * x ** 2 - 70 * x
fprime = lambda x: 4 * x ** 3 - 42 * x ** 2 + 120 * x - 70

x0 = 0
x1 = 0.1
```

```python
secant_minimizer(x0, x1, 1e-9)
```
