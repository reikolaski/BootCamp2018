```python
import numpy as np
from matplotlib import pyplot as plt
```

### Exercise 8.1

```python
x = np.linspace(-1, 5, 100)
y = np.linspace(-1, 6, 100)
X, Y = np.meshgrid(x, y)
y1 = lambda x: (2 * x + 4) / 3
y2 = lambda x: (x - 1) / 6
y3 = lambda x: 6 - x
z = lambda x, y: 5 * x - 4 * y
plt.plot(x, y1(x), c='r')
plt.plot(x, y2(x), c='r')
plt.plot(x, y3(x), c='r')
plt.xlim(0, 5)
plt.ylim(0, 6)
plt.fill_between(x, y1(x), y3(x), where=(y3(x) >= y1(x)), color='gray')
plt.contour(X, Y, z(X, Y))
plt.xlabel("x_1")
plt.ylabel("x_2")
plt.show()
```

Optimal point: $(\frac{14}{5}, \frac{16}{5})$

```python
z(14 / 5, 16 / 5)
```

Optimum value: $1.2$

### Exercise 8.2(i)

```python
x = np.linspace(-1, 12, 100)
y = np.linspace(-1, 10, 100)
X, Y = np.meshgrid(x, y)
y1 = lambda x: (15 - x) / 3
y2 = lambda x: (18 - 2 * x) / 3
y3 = lambda x: x - 4
z = lambda x, y: 3 * x + y
plt.plot(x, y1(x), c='r')
plt.plot(x, y2(x), c='r')
plt.plot(x, y3(x), c='r')
plt.xlim(0, 12)
plt.ylim(0, 10)
plt.fill_between(x, y1(x), where=(y2(x) >= y1(x)), interpolate=True, color='gray')
plt.fill_between(x, y3(x), y2(x), where=((y2(x) <= y1(x)) & (y2(x) >= y3(x))), interpolate=True, color='gray')
plt.contour(X, Y, z(X, Y))
plt.xlabel("x_1")
plt.ylabel("x_2")
plt.show()
```

Optimal point: $(6, 2)$

```python
z(6, 2)
```

Optimum value: $20$

### Exercise 8.2(ii)

```python
x = np.linspace(-1, 30, 100)
y = np.linspace(-1, 40, 100)
X, Y = np.meshgrid(x, y)
y1 = lambda x: x + 11
y2 = lambda x: 27 - x
y3 = lambda x: (90 - 2 * x) / 5
z = lambda x, y: 4 * x + 6 * y
plt.plot(x, y1(x), c='r')
plt.plot(x, y2(x), c='r')
plt.plot(x, y3(x), c='r')
plt.xlim(0, 30)
plt.ylim(0, 40)
plt.fill_between(x, y1(x), where=(y3(x) >= y1(x)), interpolate=True, color='gray')
plt.fill_between(x, y3(x), where=((y3(x) <= y2(x)) & (y3(x) < y1(x))), interpolate=True, color='gray')
plt.fill_between(x, y2(x), where=(y2(x) <= y3(x)), color='gray')
plt.contour(X, Y, z(X, Y))
plt.xlabel("x_1")
plt.ylabel("x_2")
plt.show()
```

Optimal point: $(15, 12)$

```python
z(15, 12)
```

Optimum value: $132$
